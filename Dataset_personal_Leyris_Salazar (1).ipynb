```python
import pandas as pd

# carga archivo CSV
df = pd.read_csv("Spotify most streamed.csv")

print("primeras 5 filas")
display(df.head())

print("ultima 5 filas:")
display(df.tail())

print("dimenciones (filas, columnas):")
print(df.shape)

print("informacion general:")
df.info()

print("estadisticas descriptivas:")
display(df.describe(include='all'))



```

```python
print("Columnas disponibles en el dataset:")
print(df.columns)

df.columns = df.columns.str.strip()
df['Streams'] = df['Streams'].replace(',', '', regex=True)
df['Streams'] = pd.to_numeric(df['Streams'], errors='coerce')

print("\n estadisticas interesantes:")

avg_streams = df['Streams'].mean()

top_song = df.loc[df['Streams'].idxmax(), 'Artist and Title']

df['Artist'] = df['Artist and Title'].apply(lambda x: x.split(' - ')[0])
top_artist = df['Artist'].mode()[0]

print(f"Promedio de reproducciones: {avg_streams:,.0f}")
print(f"Canción más escuchada: {top_song}")
print(f"Artista con más canciones en el ranking: {top_artist}")
```

```python
## NUEVA COLUMNA
df['Streams (millions)'] = df['Streams'] / 1_000_000
print("\n Nueva columna 'Streams (millions)' creada correctamente.")
display(df.head())


```

```python
## HALLAZGOS
print("\n Hallazgos del Análisis:")
print(f"1️ El dataset contiene información sobre canciones populares en Spotify.")
print(f"2️Tiene {df.shape[0]} canciones y {df.shape[1]} columnas.")
print(f"3️ El promedio de reproducciones es de aproximadamente {avg_streams:,.0f} streams.")
print(f"4️ La canción más reproducida es '{top_song}'.")
print(f"5️ El artista con más canciones en el top es '{top_artist}'.")
print(f"6️ Se agregó una nueva columna que muestra las reproducciones en millones.")
print(f"7️ Este análisis ayuda a conocer tendencias musicales globales.")
```
