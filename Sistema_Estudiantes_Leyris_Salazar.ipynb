```python
estudiantes = []
def agregar_estudiante():
    nombre = input("Nombre del estudiante: ")
    edad = input("Edad: ")
    grado = input("Grado: ")
    materias = input("Materias favoritas (separadas por comas): ").split(",")

    estudiante = {
        "nombre": nombre,
        "edad": edad,
        "grado": grado,
        "materias_favoritas": [m.strip() for m in materias]
    }

    estudiantes.append(estudiante)
    print(f"\nâœ… Estudiante {nombre} agregado con Ã©xito.\n")

# ------------------------------
# FunciÃ³n para mostrar estudiantes
# ------------------------------
def mostrar_estudiantes():
    if not estudiantes:
        print("\nâš  No hay estudiantes registrados.\n")
        return

    print("\nğŸ“‹ Lista de Estudiantes:\n")
    for i, est in enumerate(estudiantes, start=1):
        print(f"{i}. Nombre: {est['nombre']}")
        print(f"   Edad: {est['edad']}")
        print(f"   Grado: {est['grado']}")
        print(f"   Materias Favoritas: {', '.join(est['materias_favoritas'])}\n")

# ------------------------------
# FunciÃ³n para eliminar estudiante
# ------------------------------
def eliminar_estudiante():
    mostrar_estudiantes()
    nombre = input("Nombre del estudiante a eliminar: ")

    for est in estudiantes:
        if est['nombre'].lower() == nombre.lower():
            estudiantes.remove(est)
            print(f"\nğŸ—‘ Estudiante {nombre} eliminado.\n")
            return
    print("\nâŒ Estudiante no encontrado.\n")

# ------------------------------
# FunciÃ³n para actualizar informaciÃ³n
# ------------------------------
def actualizar_estudiante():
    mostrar_estudiantes()
    nombre = input("Nombre del estudiante a actualizar: ")

    for est in estudiantes:
        if est['nombre'].lower() == nombre.lower():
            print("\nÂ¿QuÃ© deseas actualizar?")
            print("1. Nombre")
            print("2. Edad")
            print("3. Grado")
            print("4. Materias favoritas")
            opcion = input("Elige una opciÃ³n: ")

            if opcion == "1":
                est["nombre"] = input("Nuevo nombre: ")
            elif opcion == "2":
                est["edad"] = input("Nueva edad: ")
            elif opcion == "3":
                est["grado"] = input("Nuevo grado: ")
            elif opcion == "4":
                est["materias_favoritas"] = input("Nuevas materias (separadas por comas): ").split(",")
            else:
                print("âš  OpciÃ³n no vÃ¡lida.")
                return
            print("\nâœ… InformaciÃ³n actualizada correctamente.\n")
            return

    print("\nâŒ Estudiante no encontrado.\n")

# ------------------------------
# Agregar 5 estudiantes de ejemplo
# ------------------------------
estudiantes = [
    {"nombre": "Ana LÃ³pez", "edad": "15", "grado": "10", "materias_favoritas": ["MatemÃ¡ticas", "Arte"]},
    {"nombre": "Carlos PÃ©rez", "edad": "16", "grado": "11", "materias_favoritas": ["Ciencias", "Deportes"]},
    {"nombre": "LucÃ­a Torres", "edad": "14", "grado": "9", "materias_favoritas": ["Historia", "MÃºsica"]},
    {"nombre": "Pedro GÃ³mez", "edad": "17", "grado": "12", "materias_favoritas": ["InglÃ©s", "ComputaciÃ³n"]},
    {"nombre": "SofÃ­a Rivera", "edad": "15", "grado": "10", "materias_favoritas": ["Literatura", "Arte"]}
]

# ------------------------------
# MenÃº Interactivo
# ------------------------------
def menu():
    while True:
        print("========= MENÃš DEL SISTEMA DE ESTUDIANTES =========")
        print("1. Agregar estudiante")
        print("2. Mostrar estudiantes")
        print("3. Eliminar estudiante")
        print("4. Actualizar informaciÃ³n")
        print("5. Salir")
        opcion = input("Elige una opciÃ³n: ")

        if opcion == "1":
            agregar_estudiante()
        elif opcion == "2":
            mostrar_estudiantes()
        elif opcion == "3":
            eliminar_estudiante()
        elif opcion == "4":
            actualizar_estudiante()
        elif opcion == "5":
            print("\nğŸ‘‹ Saliendo del sistema. Â¡Hasta luego!")
            break
        else:
            print("\nâš  OpciÃ³n no vÃ¡lida, intenta de nuevo.\n")
```
