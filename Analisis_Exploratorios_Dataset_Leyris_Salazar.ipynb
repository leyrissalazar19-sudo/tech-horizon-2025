```python
import pandas as pd
df = pd.read_csv("MoviesOnStreamingPlatforms.csv")
#mostrar info
print("shape:", df.shape)
df.head(8)
```

```python
#eliminar columnas que no necesito
df = df.drop(columns=["Unnamed: 0"])
#ver los nulos
print("Valores nulos por columnas:")
print(df.isnull().sum())

df["Rotten Tomatoes"] = df["Rotten Tomatoes"].astype(str).str.replace("%", "")
df["Rotten Tomatoes"] = pd.to_numeric(df["Rotten Tomatoes"], errors="coerce")

df["Year"] = pd.to_numeric(df["Year"], errors="coerce")

df.head() #mostrar primeras filas limpias
```

```python
#COLUMNA NUEVA
df["NumPlatforms"] = df[["Netflix", "Hulu", "Prime Video", "Disney+"]].sum(axis=1)
df["IsOnNetflix"] = df["Netflix"]

df[["Title", "Year", "Rotten Tomatoes", "NumPlatforms", "IsOnNetflix"]].head()
```

```python
#Promedio anual de Rotten Tomatoes por ao
promedio_plataformas = df.groupby("Year")["NumPlatforms"].mean()
print(promedio_plataformas)
```

```python
#aplicar 3 filtros diferentes
# 1. Pelicula despues de 2015
filtro1 = df[(df["Type"] == "Movie") & (df["Year"] > 2015)]
print("Pelculas despus de 2015:", filtro1.shape)
display(filtro1.head())

# 2. Rotten Tomatoes mayor o igual a 80
filtro2 = df[df["Rotten Tomatoes"] >= 80]
print("Pelculas con Rotten Tomatoes >= 80:", filtro2.shape)
display(filtro2.head())

# 3. Contenido que solo est en una plataforma
filtro3 = df[df["NumPlatforms"] == 1]
print("Contenido exclusivo (una plataforma):", filtro3.shape)
display(filtro3.head())
```
